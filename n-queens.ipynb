```python
from typing import List

class NQueen:

    n: int
    grid: List[List[int]]

    def __init__(self, n):
        self.n = n
        self.create_grid(n)

    def create_grid(self, n: int):
        grid = [0] * n

        for i in range(n):
            grid[i] = [0] * n
            for j in range(n):
                grid[i][j] = 0

        self.grid = grid

    def print_grid(self):
        print('--------------->')
        print('grid:')
        for i in range(self.n):
            print(self.grid[i])
        print('--------------->')

    def place_queen(self, row: int, col: int):
        self.grid[row][col] = 1
        print('--------------->')
        print('placed queen at x=', row, " y=", col)
        self.print_grid()
        print('--------------->')

    def remove_queen(self, row: int, col: int):
        self.grid[row][col] = 0
        print('--------------->')
        print('removed queen at x=', row, " y=", col)
        print('grid:')
        self.print_grid()
        print('--------------->')

    def under_attack(self, row: int, col: int) -> bool:
        rows: List[int] = []
        cols: List[int] = []
        dias: List[int] = []

        print(self.grid)

        for i in range(self.n):
            rows.append(self.grid[row][i])
            cols.append(self.grid[i][col])

        x = 0
        while x < self.n:
            dias.append(self.grid[x][x])
            x += 1

        if 1 in rows or 1 in cols or 1 in dias:
            print('--------------->')
            print('rows: ', rows, 'cols: ', cols, 'dias',  dias, 'is_not_under_attack', True)
            print('--------------->')
            self.print_grid()
            return True
        else:
            print('--------------->')
            print('rows: ', rows, 'cols: ', cols, 'dias',  dias, 'is_not_under_attack', False)
            print('--------------->')
            return False

    def back_track(self, row: int, count: int) -> int:
        for col in range(self.n):
            print('--------------->')
            print('row: ', row, 'col: ', col)
            print('--------------->')
            if not self.under_attack(row, col):
                self.place_queen(row, col)
                next_row = row + 1
                if next_row == self.n:
                    count += 1
                else:
                    print('--------------->')
                    print('back tracking:', 'next_row :', next_row, 'count', count)
                    print('--------------->')
                    count = self.back_track(next_row, count)
                self.remove_queen(row, col)

        return count

    def find_n_queen(self):
        count = self.back_track(0, 0)

        print('--------------->')
        print('FINAL count: ', count)
        print('--------------->')

nQueen = NQueen(4)
nQueen.find_n_queen()
```
